```python
from datetime import datetime
import pytz
import json


tz_ofs = []

for tz in pytz.common_timezones:
    now = datetime.now(pytz.timezone(tz))
    ofs = now.strftime("%z")
    tz_ofs.append((int(ofs), tz))
print(tz_ofs)
print(json.dumps(tz_ofs))
```

```python
tz_ofs_dict = {}
for tz_of in tz_ofs:
    tz_ofs_dict[tz_of[0]] = tz_of[1]
print(tz_ofs_dict)
```

```python
print(json.dumps(tz_ofs_dict))
```
