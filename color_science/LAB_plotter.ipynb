```python
import colour
import numpy as np
from colour.plotting import chromaticity_diagram_plot_CIE1931
from matplotlib import pylab


Lab = np.array([76.90, -5.53, 2.47])
XYZ = colour.Lab_to_XYZ(Lab)
xy = colour.XYZ_to_xy(XYZ)
x, y = xy
x = round(x, 4)
y = round(y, 4)

chromaticity_diagram_plot_CIE1931(standalone=False, figure_size=(10, 10), grid=False, bounding_box=[0, 0.85, 0, 0.9])
pylab.plot(x, y, 'o', markersize=7, markeredgewidth=1, markerfacecolor="None", markeredgecolor='black')
pylab.annotate((('%.4f' % x), ('%.4f' % y)), xy=xy, xytext=(-50, 30), 
               textcoords='offset points', 
               arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=-0.2'))
```

```python
from colour.plotting import single_colour_swatch_plot, ColourSwatch


print(XYZ)
RGB = colour.XYZ_to_sRGB(XYZ)
print(RGB)
single_colour_swatch_plot(ColourSwatch('Sample', RGB), text_size=32)
```
